# Feature Selection

Feature Selection is one of the core concepts in machine learning which hugely impacts the performance of your model. 
The data features that you use to train your machine learning models have a huge influence on the performance you can achieve.

# Univariate Selection

Statistical tests can be used to select those features that have the strongest 
relationship with the output variable.

scikit-learn library provides the SelectKBest class that can be 
used with a suite of different statistical tests to select a specific number of features.

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_regression
```

```python
data = pd.read_csv('phone_train.csv')
```

```python
data.head()
```

```python
X = data.iloc[:,0:20]  #independent columns
y = data.iloc[:,-1]    #target column i.e price range
```

```python
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
```

```python
fit = bestfeatures.fit(X,y)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
dfscores
```

```python
dfcolumns
```

```python
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores
```

```python
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

# Feature Importance

Feature importance is an inbuilt class that comes with Tree Based Classifiers, we will be using Extra Tree Classifier for extracting 
the top 10 features for the dataset.

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
```

```python
# Y=price range
# x=rest other features
```

```python
model = ExtraTreesClassifier()
model.fit(X,y)
```

```python
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
```

```python
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
```

```python
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```
