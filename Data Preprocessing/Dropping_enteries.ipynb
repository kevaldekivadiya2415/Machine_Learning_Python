```python
# Create a dataframe
data1 = {'name': ['Ajay', 'Aakash', 'Tina', 'Jaya', 'Aamir'], 
        'year': [2012, 2012, 2013, 2014, 2014], 
        'sales': [4, 24, 31, 2, 3]}
df = pd.DataFrame(data1, index = ['Delhi', 'Mumbai', 'Bangalore', 'Chennai', 'Pune'])
df
```

```python
new_df=df.set_index('year')
new_df
```

```python
new_df.drop([2012])
```

```python
# Drop an observation (row)
df.drop(['Delhi', 'Pune'])
```

```python
# Drop a variable (column)
#Note: axis=1 denotes that we are referring to a column, not a row
df.drop('sales', axis=1)
```

```python
# Drop a row if it contains a certain value (in this case, “Tina”)
# Specifically: Create a new dataframe called df that includes all rows where the value of a 
# cell in the name column does not equal “Tina”
df[df.name !='Tina']
```

```python
# Drop a row by row number (in this case, row 3)
# Note that Pandas uses zero based numbering, so 0 is the first row, 1 is the second row, etc.
df.drop(df.index[2])
```

```python
# can be extended to dropping a range
df.drop(df.index[[2,3]])
```

```python
# or dropping relative to the end of the DF.
df.drop(df.index[-2])
```

```python
# we can select ranges relative to the top or drop relative to the bottom of the DF as well.
# keep top 3
df[:3]
```

```python
#drop bottom 3
df[:-3] 
```

# Drop Columns from a Dataframe using iloc[] and drop() method.

Remove all columns between a specific column to another columns.

```python
# Remove all columns between column index 1 to 3 
data.drop(data.iloc[:,1:3],inplace=True, axis=1)
data.drop(data.iloc[:,1:3],axis=1).head()
```

# Drop Columns from a Dataframe using loc[] and drop() method.

Remove all columns between a specific column name to another columns name.

```python
# # Remove all columns between column name 'Country' to 'Total_Profit'
```
